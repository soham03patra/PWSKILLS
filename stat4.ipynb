## Q(1)

Both t-tests and z-tests are statistical methods used to make inferences about population parameters based on sample data. However, they differ in their application and assumptions, particularly in relation to sample size and the known variance of the population.

1. T-Test:

Use Case:
T-tests are generally used when the sample size is small (typically less than 30) and the population standard deviation is unknown.

Example :

Suppose you want to test whether there is a significant difference in the mean scores of two groups (e.g., Group A and Group B) on a particular test. You collect samples from each group, and since the population standard deviation is unknown, you use a t-test.

2. Z-Test:

Use Case:

Z-tests are appropriate when the sample size is large (typically greater than 30) and the population standard deviation is known.

Example :

Imagine you are conducting quality control for a manufacturing process, and you want to test whether the mean weight of a product matches the specified target weight. If you have a large sample size and know the population standard deviation (perhaps from historical data), you might use a z-test.

## Q(4)

Bayes's theorem is a mathematical formula that allows us to update the probability of an event based on new information or evidence.
It's often used to answer questions about conditional probability, which is the probability of one event happening given that another event has already occurred.
Formula:

P(A|B) = (P(B|A) * P(A)) / P(B)

P(A|B): The probability of event A happening given that event B has already happened (posterior probability)
P(B|A): The probability of event B happening given that event A has already happened (likelihood)
P(A): The probability of event A happening, regardless of event B (prior probability)
P(B): The probability of event B happening, regardless of event A (marginal probability)
Example:

Scenario: Imagine a medical test that's 99% accurate in detecting a disease when the disease is present (true positive rate), but also has a 1% false positive rate (meaning it incorrectly indicates the disease in 1% of healthy people). If a person tests positive, what's the probability they actually have the disease?

Applying Bayes's theorem:

P(Disease|Positive) = (P(Positive|Disease) * P(Disease)) / P(Positive)
P(Positive|Disease) = 0.99 (true positive rate)
P(Disease) = 0.001 (assume a rare disease with 0.1% prevalence)
P(Positive) = P(Positive|Disease) * P(Disease) + P(Positive|No Disease) * P(No Disease)
P(Positive) = 0.99 * 0.001 + 0.01 * 0.999 = 0.01098
P(Disease|Positive) = (0.99 * 0.001) / 0.01098 ≈ 0.0902

To calculate the 95% confidence interval for a sample with a mean of 50 and a standard deviation of 5, we'll use the formula for the confidence interval for the population mean (\(\mu\)):

\[ \text{Confidence Interval} = \bar{X} \pm Z \left( \frac{\sigma}{\sqrt{n}} \right) \]

Where:
- \(\bar{X}\) is the sample mean.
- \(Z\) is the critical value from the standard normal distribution corresponding to the desired confidence level. For a 95% confidence interval, \(Z \approx 1.96\).
- \(\sigma\) is the population standard deviation.
- \(n\) is the sample size.

Given:
- \(\bar{X} = 50\) (sample mean).
- \(\sigma = 5\) (population standard deviation).
- \(Z \approx 1.96\) (for a 95% confidence interval).

Let's calculate the confidence interval:

\[ \text{Confidence Interval} = 50 \pm 1.96 \left( \frac{5}{\sqrt{n}} \right) \]

Now, let's interpret the results. If you have a specific sample size (\(n\)), you can substitute that value into the formula. If you don't have a specific sample size, you won't be able to provide a specific interval, but you can discuss the concept.

For example, if \(n = 25\):

\[ \text{Confidence Interval} = 50 \pm 1.96 \left( \frac{5}{\sqrt{25}} \right) \]

\[ \text{Confidence Interval} = 50 \pm 1.96 \left( \frac{5}{5} \right) \]

\[ \text{Confidence Interval} = 50 \pm 1.96 \times 1 \]

\[ \text{Confidence Interval} \approx (48.04, 51.96) \]

Interpretation:
"We are 95% confident that the true population mean (\(\mu\)) falls within the interval (48.04, 51.96). This means that if we were to take many random samples and compute a 95% confidence interval for each sample, we would expect approximately 95% of those intervals to contain the true population mean."

## Q(8)

**Margin of Error in a Confidence Interval:**
The margin of error (MOE) in a confidence interval is a measure of the precision or uncertainty associated with the estimation of a population parameter. It represents the range around a point estimate within which we are reasonably confident the true parameter lies. The margin of error is influenced by the variability in the sample data and the desired level of confidence.

**Formula for Margin of Error:**
\[ \text{Margin of Error} = Z \times \left( \frac{\sigma}{\sqrt{n}} \right) \]

Where:
- \( Z \) is the critical value from the standard normal distribution or t-distribution, depending on the sample size and desired confidence level.
- \( \sigma \) is the population standard deviation (or an estimate of it).
- \( n \) is the sample size.

**Effect of Sample Size on Margin of Error:**
As the sample size (\( n \)) increases, the margin of error decreases. In other words, larger sample sizes lead to more precise estimates, resulting in a narrower confidence interval. This relationship is due to the denominator in the formula (\( \sqrt{n} \)). As \( n \) increases, the square root of \( n \) increases, leading to a reduction in the margin of error.

**Example Scenario:**
Suppose you are conducting a survey to estimate the average income of a certain population. You decide to use a 95% confidence level, and you want to determine the sample size needed to achieve a certain margin of error. You estimate the population standard deviation to be $10,000.

If you initially plan to survey 100 individuals (\( n = 100 \)), the margin of error might be, for example, $2,000. However, if you decide to increase the sample size to 400 individuals (( n = 400 )), the margin of error might decrease to $1,000.

**Illustrative Calculation:**
\[ \text{Margin of Error}_1 = Z \times \left( \frac{\sigma}{\sqrt{100}} \right) \]
\[ \text{Margin of Error}_2 = Z \times \left( \frac{\sigma}{\sqrt{400}} \right) \]

In this example, \( \text{Margin of Error}_2 \) is smaller than \( \text{Margin of Error}_1 \) due to the larger sample size (\( n = 400 \)), resulting in a more precise estimate of the average income.

In summary, a larger sample size contributes to a smaller margin of error, indicating increased precision in estimating population parameters.

\[ \text{Confidence Interval} = \hat{p} \pm Z \times \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} \]

where:
- \(\hat{p}\) is the sample proportion,
- \(Z\) is the critical value from the standard normal distribution (for a 95% confidence interval, \(Z \approx 1.96\)),
- \(n\) is the sample size.

Given:
- \(\hat{p} = 0.65\) (sample proportion),
- \(Z \approx 1.96\) (for a 95% confidence interval),
- \(n = 500\) (sample size).

Let's substitute these values into the formula:

\[ \text{Confidence Interval} = 0.65 \pm 1.96 \times \sqrt{\frac{0.65 \times (1 - 0.65)}{500}} \]

Now, calculate the values:

\[ \text{Confidence Interval} = 0.65 \pm 1.96 \times \sqrt{\frac{0.65 \times 0.35}{500}} \]

\[ \text{Confidence Interval} = 0.65 \pm 1.96 \times \sqrt{\frac{0.2275}{500}} \]

\[ \text{Confidence Interval} = 0.65 \pm 1.96 \times \sqrt{0.000455} \]

\[ \text{Confidence Interval} = 0.65 \pm 1.96 \times 0.02135 \]

Now, compute the upper and lower bounds of the confidence interval:

\[ \text{Lower Bound} = 0.65 - (1.96 \times 0.02135) \]

\[ \text{Upper Bound} = 0.65 + (1.96 \times 0.02135) \]

\[ \text{Lower Bound} \approx 0.608 \]
\[ \text{Upper Bound} \approx 0.692 \]

**Conclusion:**
