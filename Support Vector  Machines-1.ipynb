# Q1. What is the mathematical formula for a linear SVM?



- A linear Support Vector Machine (SVM) aims to find a hyperplane that separates the data into two classes. The mathematical formula for a linear SVM can be expressed as follows:

- Given a set of training data points \((x_1, y_1), (x_2, y_2), ..., (x_n, y_n)\), where \(x_i\) is the input vector and \(y_i\) is the corresponding class label (\(y_i \in \{-1, 1\}\)), the goal is to find a weight vector \(w\) and a bias term \(b\) such that the hyperplane equation \(w \cdot x + b = 0\) separates the data with maximum margin.

#### The optimization problem for linear SVM can be formulated as:

\[\text{Minimize } \frac{1}{2} \|w\|^2\]
subject to \[y_i(w \cdot x_i + b) \geq 1 \text{ for } i = 1, 2, ..., n\]

- Here, \(\|w\|\) represents the Euclidean norm of the weight vector \(w\), and the inequality constraint ensures that each data point is correctly classified with a margin of at least 1.

- The solution to this optimization problem gives the optimal values for \(w\) and \(b\) that define the hyperplane separating the two classes. The SVM is particularly powerful for finding a hyperplane that maximizes the margin between classes, making it robust to outliers and effective in high-dimensional spaces.

# Q2. What is the objective function of a linear SVM?



The objective function of a linear Support Vector Machine (SVM) is to find the hyperplane that best separates the data into different classes while maximizing the margin between the classes. The hyperplane is a decision boundary that separates the data points of one class from another. The margin is defined as the distance between the hyperplane and the nearest data point from either class.

In the case of a linear SVM, the objective function is formulated to minimize the misclassification error and simultaneously maximize the margin. Mathematically, the objective function can be expressed as follows:

\[ \text{Minimize } \frac{1}{2} \|w\|^2 + C \sum_{i=1}^{N} \xi_i \]

subject to the constraints:

\[ y_i (w \cdot x_i + b) \geq 1 - \xi_i \]
\[ \xi_i \geq 0 \]

Here:
- \(w\) is the weight vector that defines the hyperplane,
- \(b\) is the bias term,
- \(x_i\) is a data point,
- \(y_i\) is the class label of \(x_i\) (\(y_i = \pm 1\)),
- \(\xi_i\) are slack variables that allow for some degree of misclassification,
- \(C\) is a regularization parameter that controls the trade-off between achieving a low training error and a large margin,
- \(N\) is the number of training examples.

The objective is to find the values of \(w\) and \(b\) that minimize the expression while satisfying the constraints. The regularization parameter \(C\) determines the balance between achieving a small margin and allowing for some misclassification. Higher values of \(C\) result in a smaller margin but fewer misclassifications, while lower values of \(C\) result in a larger margin but allow for more misclassifications.

# Q3. What is the kernel trick in SVM?



The kernel trick is a technique used in Support Vector Machines (SVMs) to handle non-linear decision boundaries by implicitly mapping the input data into a higher-dimensional space. In the original formulation of SVM, it's designed for linearly separable data with a linear decision boundary. However, many real-world problems involve non-linear relationships among the features, making linear separation insufficient.

The kernel trick allows SVMs to operate in a higher-dimensional space without explicitly calculating the coordinates of the data points in that space. Instead of computing the transformation explicitly, the kernel function is used to compute the dot product of the transformed data points without explicitly finding their coordinates in the higher-dimensional space. This is computationally more efficient, as it avoids the need to represent and store the data in the higher-dimensional space explicitly.

The most common types of kernels used in SVM include:

1. **Linear Kernel (\(K(x, y) = x \cdot y\)):** This is the standard dot product of the input features. It is equivalent to the original linear SVM.

2. **Polynomial Kernel (\(K(x, y) = (x \cdot y + c)^d\)):** This introduces non-linearity by computing the dot product raised to a power \(d\). The parameters \(c\) and \(d\) can be tuned.

3. **Radial Basis Function (RBF) or Gaussian Kernel (\(K(x, y) = \exp\left(-\frac{\|x - y\|^2}{2\sigma^2}\right)\)):** This is a popular choice for non-linear problems. It uses the Gaussian function to measure the similarity between two data points, where \(\sigma\) is a parameter controlling the width of the kernel.

4. **Sigmoid Kernel (\(K(x, y) = \tanh(\alpha x \cdot y + c)\)):** This is based on the hyperbolic tangent function and is sometimes used in neural network applications.

The kernel trick allows SVMs to capture complex relationships in the data and find non-linear decision boundaries by implicitly transforming the data into higher-dimensional spaces. It's a powerful tool for handling non-linear classification problems.

# Q4. What is the role of support vectors in SVM Explain with example 



- In Support Vector Machines (SVM), support vectors play a crucial role in determining the optimal hyperplane and defining the decision boundary. Support vectors are the data points that lie closest to the decision boundary (hyperplane) and have a direct influence on its position and orientation. These points are the most critical for determining the margin, which is the distance between the hyperplane and the nearest data point from either class.

Here's a step-by-step explanation of the role of support vectors in SVM:

1. **Optimal Hyperplane:**
   - The goal of SVM is to find the hyperplane that maximizes the margin between the two classes.
   - The margin is defined as the distance between the hyperplane and the nearest data point (support vector) from either class.

2. **Support Vectors:**
   - Support vectors are the data points that lie on the margin or are misclassified.
   - These are the critical points that determine the position and orientation of the hyperplane.

3. **Margin:**
   - The margin is maximized when the hyperplane is positioned in such a way that the distance between it and the nearest support vector from either class is maximized.

4. **Decision Boundary:**
   - The decision boundary is determined by the position and orientation of the hyperplane.
   - The hyperplane is chosen to maximize the margin while still correctly classifying the training data.

5. **Tolerance for Misclassification:**
   - SVM allows for some degree of misclassification through the introduction of slack variables (represented as \(\xi\) in the objective function).
   - Support vectors are the data points where these slack variables are non-zero.

Example:

Consider a simple 2D classification problem with two classes, +1 and -1. The training data consists of points in the following way:

- Class +1: \((1, 2), (2, 3), (3, 3)\)
- Class -1: \((2, 1), (3, 2), (4, 1)\)

In this case, the support vectors would be the points \((1, 2)\) and \((4, 1)\) from classes +1 and -1, respectively. These are the points lying on the margin or inside the margin, influencing the placement of the optimal hyperplane. The optimal hyperplane, in turn, is determined by the support vectors and separates the two classes with the maximum margin.

# Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

X = np.array([[2, 2], [3, 1], [3, 3], [0, -1], [1, 0], [1, -1]])
y = np.array([1, 1, 1, -1, -1, -1])
```

```python

clf = svm.SVC(kernel='linear', C=1)
clf.fit(X, y)
```

```python

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100)
```

```python

w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-1, 4)
yy = a * xx - (clf.intercept_[0]) / w[1]
plt.plot(xx, yy, 'k-')
```

```python

margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
yy_down = yy - np.sqrt(1 + a ** 2) * margin
yy_up = yy + np.sqrt(1 + a ** 2) * margin
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
```

```python

plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=150, facecolors='none', edgecolors='k')

plt.title('Hyperplane and Margins')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Soft Margin and Hard Margin:

```python

X_soft = np.array([[2, 2], [3, 1], [3, 3], [0, -1], [1, 0], [1, -1]])
y_soft = np.array([1, 1, 1, -1, -1, -1])
```

```python

X_hard = np.array([[2, 2], [3, 1], [3, 3], [0, -1], [1, 0], [1, -1]])
y_hard = np.array([1, 1, 1, -1, -1, -1])
```

```python

clf_soft = svm.SVC(kernel='linear', C=1)
clf_hard = svm.SVC(kernel='linear', C=1000)

clf_soft.fit(X_soft, y_soft)
clf_hard.fit(X_hard, y_hard)
```

```python

plt.scatter(X_soft[:, 0], X_soft[:, 1], c=y_soft, cmap=plt.cm.Paired, marker='o', s=100)
```

```python

w_soft = clf_soft.coef_[0]
a_soft = -w_soft[0] / w_soft[1]
yy_soft = a_soft * xx - (clf_soft.intercept_[0]) / w_soft[1]
margin_soft = 1 / np.sqrt(np.sum(clf_soft.coef_ ** 2))
yy_down_soft = yy_soft - np.sqrt(1 + a_soft ** 2) * margin_soft
yy_up_soft = yy_soft + np.sqrt(1 + a_soft ** 2) * margin_soft
plt.plot(xx, yy_soft, 'k-')
plt.plot(xx, yy_down_soft, 'k--')
plt.plot(xx, yy_up_soft, 'k--')
```

```python

w_hard = clf_hard.coef_[0]
a_hard = -w_hard[0] / w_hard[1]
yy_hard = a_hard * xx - (clf_hard.intercept_[0]) / w_hard[1]
margin_hard = 1 / np.sqrt(np.sum(clf_hard.coef_ ** 2))
yy_down_hard = yy_hard - np.sqrt(1 + a_hard ** 2) * margin_hard
yy_up_hard = yy_hard + np.sqrt(1 + a_hard ** 2) * margin_hard
plt.plot(xx, yy_hard, 'k-.')
plt.plot(xx, yy_down_hard, 'k-.')
plt.plot(xx, yy_up_hard, 'k-.')
```

```python

plt.scatter(clf_soft.support_vectors_[:, 0], clf_soft.support_vectors_[:, 1],
            s=150, facecolors='none', edgecolors='k')

plt.title('Soft Margin (C=1) vs. Hard Margin (C=1000)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Q6. SVM Implementation through Iris dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python

svm_classifier = SVC(kernel='linear', C=1, random_state=42)
```

```python

svm_classifier.fit(X_train_std, y_train)
```

```python

y_pred = svm_classifier.predict(X_test_std)
```

```python

accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python

print(f"Accuracy: {accuracy:.2f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

```python

h = 0.02
x_min, x_max = X_train_std[:, 0].min() - 1, X_train_std[:, 0].max() + 1
y_min, y_max = X_train_std[:, 1].min() - 1, X_train_std[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```
